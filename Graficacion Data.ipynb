```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df_z_countries=pd.read_csv('zika_by_countries.csv').set_index('country')
df_z_locations_brazil=pd.read_csv('zica_brasil_locations.csv').set_index('location')
df_z_contagios=pd.read_csv('contagios.csv').set_index('report_date')
```

```python
df_z_countries.plot(kind='bar',figsize=(10,7), title='Countries')
plt.xlabel('Countries')
plt.ylabel('Contagions (Mill)')
```

```python
df_z_locations_brazil.plot(kind='bar',figsize=(10,7), title='Brazil Regions')
plt.xlabel('Regions')
plt.ylabel('Contagion')
```

```python
df_z_contagios.plot(kind='bar',figsize=(10,7), title='Contagions by time')
plt.xlabel('Dates')
plt.ylabel('Contagions (Mill)')
```
